# Sélection des variables d'intérêt
1. Pour sélectionner les variables, focus sur les années 2000-2020

```python
# retrait colonnes > 2020
colonne = EdStatsData.columns
nbColonnes = len(colonne)
while colonne[nbColonnes-1] != '2020':
    nbColonnes -= 1
EdStatsData00_20 = EdStatsData.drop(colonne[nbColonnes : len(colonne)], axis=1)

# pour enlever les colonnes 1970-2000 qui sont trop anciennes (essor internet)
colonneAvant2000 = 4
while colonne[colonneAvant2000] != '2000':
    colonneAvant2000 += 1
EdStatsData00_20 = EdStatsData00_20.drop(colonne[4 : colonneAvant2000],axis = 1)

EdStatsData00_20.head(2)
```

```python
EdStatsData00_20.shape
```

```python
# proportion de NaN dans le fichier :
EdStatsData00_20.isna().mean().mean()
```

```python
# nombre d'indicateurs
EdStatsData00_20['Indicator Name'].unique().shape
```

```python
# calcul taux de remplissage de chaque ligne
EdStatsData00_20['tauxRemplissage'] = EdStatsData00_20.count(axis=1)/(EdStatsData00_20.shape[1]-4)
```

2. Catégorisation des indicateurs via la variable Topic disponible dans le fichier esStatsSeries.csv

```python
# Ajout de la colonne Topic via une jointure avec le tableau edStatSeries
indicateurs = pd.read_csv('../01_DATA/EdStatsSeries.csv')
indicateurs = indicateurs.dropna(axis = 1)
indicateurs = indicateurs.drop(['Indicator Name', 'Long definition', 'Source'], axis = 1)

EdStatsData00_20_tot = pd.merge(EdStatsData00_20, indicateurs, left_on="Indicator Code", right_on="Series Code", how='left')
EdStatsData00_20_tot.head()
EdStatsData00_20_tot = EdStatsData00_20_tot.drop('Series Code', axis=1)
```

```python
# Fonction pour récupérer la dernière date à laquelle il y a une donnée pour une ligne

def derniereDate(data):
    colonne = len(data)-1
    # tant que la case est vide ET qu'on est toujours dans les colonnes années
    while (math.isnan(data.iloc[colonne])) & (colonne>4):
        colonne -= 1
    # si on arrive au bout de la plage d'années, alors il n'y a pas de données pour cette ligne
    if ESD_8ind_countries.columns[colonne-1].startswith(('19','20')) == False:
        # si il n'y a aucune données sur la ligne
        return 'No Data'
    else:
        return data.index[colonne]


ESD_8ind_countries.insert(4, 'Dernière date renseignée', 
                            [derniereDate(ESD_8ind_countries.iloc[x]) for x in range(len(ESD_8ind_countries))])
```

```python
# pour automatiser la generation des listes de pays non renseignés par indicateurs
def listePaysNA(df,indicateur):
    indicateurPays = df[df['Indicator Name'] == indicateur]
    listeIndicateur = indicateurPays[indicateurPays['Dernière date renseignée'] == 'No Data']['Country Name'].unique()
    return listeIndicateur



# pour automatiser la génération de graph par pays
def graphPays(df, indicateur):
    indicateurPays = df[df['Indicator Name'] == indicateur]
    # suppression des pays non renseignés sur ce critère
    listeIndicateur = listePaysNA(df,indicateur)
    indicateurPays = indicateurPays.set_index('Country Name')
    indicateurPays = indicateurPays.drop(listeIndicateur, axis = 0)

    indicateurPaysT = indicateurPays.transpose()
    # retrait des lignes 1 à 4
    indicateurPaysT = indicateurPaysT[4:22]
    # conversion str en datetime
    indicateurPaysT.index = [datetime.strptime(indicateurPaysT.index[x], "%Y").year for x in range(len(indicateurPaysT))]


    plot = plt.plot(indicateurPaysT)
    plot = plt.title(indicateur)
    plot = plt.xlabel('Années')

# pour automatiser la génération de graph par regions basé sur les données pays
def graphRegions(df, indicateur, country):
    
    indicateurPays = df[df['Indicator Name'] == indicateur]
    # suppression des pays non renseignés sur ce critère
    listeIndicateur = listePaysNA(df,indicateur)
    indicateurPays = indicateurPays.set_index('Country Name')
    indicateurPays = indicateurPays.drop(listeIndicateur, axis = 0)

    # retrait des colonnes 1 à 4
    indicateurPays = indicateurPays.iloc[:,4:22]
    # conversion str en datetime
    indicateurPays.columns = [datetime.strptime(indicateurPays.columns[x], "%Y").year for x in range(indicateurPays.shape[1])]
    
    # rajout info région pour la couleur
    indicateurPays = indicateurPays.reset_index()

    indicateurPaysTOT = pd.melt(indicateurPays, id_vars = ['Country Name'], 
                                value_vars=indicateurPays.columns.values[3:],
                                var_name='Années', value_name=indicateur)

    indicateurPaysTOT = pd.merge(indicateurPaysTOT, country, left_on="Country Name", right_on="Table Name", how='left')
    indicateurPaysTOT = indicateurPaysTOT.drop(['Unnamed: 0', 'Country Code','Short Name','Long Name', 'Table Name'], axis=1)
    
    plot = sns.lineplot(data=indicateurPaysTOT, 
                        x='Années', 
                        y=indicateur, 
                        hue='Region', 
#                         estimator='mean', 
#                         ci='sd'
                   )
    plot = plt.title(indicateur)
    plot = plt.legend(bbox_to_anchor=(1.02, 1))
    plot = plt.ylabel('')
```

L'indice peut être calculé pour chaque année, mais l'année de référence ne pourra pas être 2016, du fait du manque de données (-50%)

```python
# mise à jour de l'index
ESD_8ind_countries_short = ESD_8ind_countries_short.reset_index()
ESD_8ind_countries_short = ESD_8ind_countries_short.drop('index', axis=1)
ESD_8ind_countries_short.tail()
```

### Calcul de la population scolarisée au lycée et dans le supérieur = grossEnrolment x Pop in age

```python
for pays in ESD_8ind_countries_short['Country Name'].unique():
    # pour chaque pays
    # initialisation ligne pour la population scolarisée au lycée
    rowLycee=[]
    rowLycee.append(pays)
    rowLycee.append('NaN')
    rowLycee.append('Pop scolarisée lycée')
    rowLycee.append('NaN')
    rowLycee.append('NaN')
    rowLycee.append('NaN')
    
    # initialisation ligne pour la population scolarisée en supérieur
    rowSup=[]
    rowSup.append(pays)
    rowSup.append('NaN')
    rowSup.append('Pop scolarisée Supérieur')
    rowSup.append('NaN')
    rowSup.append('NaN')
    rowSup.append('NaN')
    
    filtrePays = ESD_8ind_countries_short['Country Name'] == pays

    for i in range(6,ESD_8ind_countries_short.shape[1]):
        # pour chaque colonne année, récupère les indicateurs gross enrolment et pop in age
        # Lycée
        filtreGElycee = ESD_8ind_countries_short['Indicator Name'] == 'Gross enrolment ratio, upper secondary, both sexes (%)'
        filtrePopLycee = ESD_8ind_countries_short['Indicator Name'] == 'Population of the official age for upper secondary education, both sexes (number)'
        GELyc = ESD_8ind_countries_short[filtrePays & filtreGElycee].iloc[:,i].values
        PopLyc = ESD_8ind_countries_short[filtrePays & filtrePopLycee].iloc[:,i].values
        
        nbScoLyc = GELyc*PopLyc/100
        
        rowLycee.append(nbScoLyc[0])
        
        # Supérieur
        filtreGESup = ESD_8ind_countries_short['Indicator Name'] == 'Gross enrolment ratio, tertiary, both sexes (%)'
        filtrePopSup = ESD_8ind_countries_short['Indicator Name'] == 'Population of the official age for tertiary education, both sexes (number)'
        GESup = ESD_8ind_countries_short[filtrePays & filtreGESup].iloc[:,i].values
        PopSup = ESD_8ind_countries_short[filtrePays & filtrePopSup].iloc[:,i].values
        
        nbScoSup = GESup*PopSup/100
        
        rowSup.append(nbScoSup[0])
        
    loc = ESD_8ind_countries_short.shape[0]
    ESD_8ind_countries_short.loc[loc] = rowLycee
    ESD_8ind_countries_short.loc[loc+1] = rowSup

ESD_8ind_countries_short.tail()
```

### liste des valeurs min et max par indicateur et année tous pays confondus

```python
# récupération des info pour nommer les colonnes du df
listeAnnees = ESD_8ind_countries_short.columns[6:]
titreColonnes = listeAnnees.insert(0,'Indicateur')
i=0
minValue=[]
maxValue=[]
for indic in ESD_8ind_countries_short['Indicator Name'].unique():
    # pour chaque indicateur
    rowMin=[]
    rowMin.append(indic)
    rowMax=[]
    rowMax.append(indic)
    for Annee in listeAnnees:
        # pour chaque année, récupère la valeur min et max tout pays confondu
        rowMin.append(ESD_8ind_countries_short[ESD_8ind_countries_short['Indicator Name'] == indic][Annee].min(axis=0))
        rowMax.append(ESD_8ind_countries_short[ESD_8ind_countries_short['Indicator Name'] == indic][Annee].max(axis=0))
    minValue.append(rowMin)
    maxValue.append(rowMax)
minValue = pd.DataFrame(minValue, columns=titreColonnes)
maxValue = pd.DataFrame(maxValue, columns=titreColonnes)
```

```python
minValue.tail(2)
```

#### Calcul indice intermédiaire Pop scolarisée Supérieur

```python
indic = 'Pop scolarisée Supérieur'
listeAnnees = ESD_8ind_countries_short.columns[6:]
titreColonnes = listeAnnees.insert(0,'pays')
indiceScoSup = pd.DataFrame(columns = titreColonnes)

for pays in ESD_8ind_countries_short['Country Name'].unique():
    row=[]
    row.append(pays)
    for Annee in listeAnnees:
        valeurReelle = ESD_8ind_countries_short[(ESD_8ind_countries_short['Country Name'] == pays) & (ESD_8ind_countries_short['Indicator Name'] == indic)][Annee].values[0]
        valeurMin = minValue[minValue['Indicateur'] == indic][Annee].values[0] 
        valeurMax = maxValue[maxValue['Indicateur'] == indic][Annee].values[0]
        Indice = (valeurReelle - valeurMin)/(valeurMax - valeurMin)
        row.append(Indice)
    loc = indiceScoSup.shape[0]
    indiceScoSup.loc[loc] = row
indiceScoSup.tail()
```

#### Calcul indice intermédiaire Pop scolarisée Lycée

```python
indic = 'Pop scolarisée lycée'
listeAnnees = ESD_8ind_countries_short.columns[6:]
titreColonnes = listeAnnees.insert(0,'pays')
indiceScoLyc = pd.DataFrame(columns = titreColonnes)

for pays in ESD_8ind_countries_short['Country Name'].unique():
    row=[]
    row.append(pays)
    for Annee in listeAnnees:
        valeurReelle = ESD_8ind_countries_short[(ESD_8ind_countries_short['Country Name'] == pays) & (ESD_8ind_countries_short['Indicator Name'] == indic)][Annee].values[0]
        valeurMin = minValue[minValue['Indicateur'] == indic][Annee].values[0] 
        valeurMax = maxValue[maxValue['Indicateur'] == indic][Annee].values[0]
        Indice = (valeurReelle - valeurMin)/(valeurMax - valeurMin)
        row.append(Indice)
    loc = indiceScoLyc.shape[0]
    indiceScoLyc.loc[loc] = row
indiceScoLyc.tail()
```

### Calcul indice global scolarisation

#### tableau du maximum de la moyenne geometrique des 2 indices scolaires

```python
# récupération des info pour nommer les colonnes du df
listeAnnees = ESD_8ind_countries_short.columns[6:]
titreColonnes = listeAnnees.insert(0,'pays')
moySco = pd.DataFrame(columns = titreColonnes)

# creation du tableau avec les moyennes geometrique pour chaque pays et années
for pays in ESD_8ind_countries_short['Country Name'].unique():
    row=[]
    row.append(pays)
    for Annee in listeAnnees:
        # pour chaque année, calcul de la moyenne géométrique
        indScoLyc = indiceScoLyc[indiceScoLyc['pays'] == pays][Annee].values[0]
        indScoSup = indiceScoSup[indiceScoSup['pays'] == pays][Annee].values[0]
        moyGeo = np.sqrt(indScoLyc * indScoSup)
        row.append(moyGeo)
    loc = moySco.shape[0]
    moySco.loc[loc] = row

# creation du tableau du max de cette moyenne geometrique par annees
maxValueSco = pd.DataFrame(moySco.max(axis=0)).transpose().drop('pays', axis=1)

maxValueSco
```

#### Indice global sur la scolarisation

```python
listeAnnees = ESD_8ind_countries_short.columns[6:]
titreColonnes = listeAnnees.insert(0,'pays')
indiceSco = pd.DataFrame(columns = titreColonnes)

for pays in moySco['pays'].unique():
    row=[]
    row.append(pays)
    for Annee in listeAnnees:
        valeurReelle = moySco[moySco['pays'] == pays][Annee].values[0]
        valeurMax = maxValueSco[Annee].values[0]
        Indice = (valeurReelle - 0)/(valeurMax - 0)
        row.append(Indice)
    loc = indiceSco.shape[0]
    indiceSco.loc[loc] = row
indiceSco.head()
```

#### Indice global sur Internet

#### Indice global PIB

```python
indic = 'GDP per capita (current US$)'
listeAnnees = ESD_8ind_countries_short.columns[6:]
titreColonnes = listeAnnees.insert(0,'pays')
indicePIB = pd.DataFrame(columns = titreColonnes)

for pays in ESD_8ind_countries_short['Country Name'].unique():
    row=[]
    row.append(pays)
    for Annee in listeAnnees:
        valeurReelle = ESD_8ind_countries_short[(ESD_8ind_countries_short['Country Name'] == pays) & (ESD_8ind_countries_short['Indicator Name'] == indic)][Annee].values[0]
        valeurMin = minValue[minValue['Indicateur'] == indic][Annee].values[0] 
        valeurMax = maxValue[maxValue['Indicateur'] == indic][Annee].values[0]
        Indice = (np.log(valeurReelle) - np.log(valeurMin))/(np.log(valeurMax) - np.log(valeurMin))
        row.append(Indice)
    loc = indicePIB.shape[0]
    indicePIB.loc[loc] = row
indicePIB.tail()
```

### Calcul indice global

```python
listeAnnees = ESD_8ind_countries_short.columns[6:]
titreColonnes = listeAnnees.insert(0,'pays')
indiceGlobal = pd.DataFrame(columns = titreColonnes)

for pays in ESD_8ind_countries_short['Country Name'].unique():
    row=[]
    row.append(pays)
    for Annee in listeAnnees:
        indPIB = indicePIB[indicePIB['pays'] == pays][Annee].values[0]
        indNet = indiceNet[indiceNet['pays'] == pays][Annee].values[0]
        indSco = indiceSco[indiceSco['pays'] == pays][Annee].values[0]
        Indice = np.cbrt(indPIB * indNet * indSco)
        row.append(Indice)
    loc = indiceGlobal.shape[0]
    indiceGlobal.loc[loc] = row
indiceGlobal.head()
```
